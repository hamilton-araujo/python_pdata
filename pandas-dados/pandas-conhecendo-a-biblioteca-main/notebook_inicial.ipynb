# Manipulando os dados

## Criando colunas numéricas

```python
dados = pd.read_csv("base-de-dados/aluguel.csv", sep=';')
dados
```

```python
dados["Valor por mês"] = dados["Valor"] + dados["Condominio"]
dados
```

```python
dados.fillna(0,inplace = True)
dados['Valor por mês'] = dados['Valor'] + dados['Condominio']
dados
```

```python
dados["Valor por ano"] = dados["Valor por mês"] * 12 + dados["IPTU"]
dados
```

## Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados
```

```python
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
```

```python
data = pd.read_csv("desafios/alunos.csv")
data
```

```python
data.fillna(0,inplace=True)
data
```

```python
data["Pontos Extras"] = data["Notas"].apply(lambda x: x*0.4)
data
```

```python
data["Notas Finais"] = data["Notas"] + data["Pontos Extras"]
data
```

```python
data["Aprovado_Final"] = data["Notas Finais"].apply(lambda x: True if x >=6 else False)
data
```

```python
salvos = data.query('Aprovado_Final == True and Aprovado == False')
salvos.to_csv('alunos_salvos.csv', index=False)
```
