```python
def corrigir_respostas():

    respostas_aluno = {}
    pontuacao = 0

    print("--- Correção de Prova ---")
    print("Digite suas respostas (A, B, C ou D) para cada questão.\n")

    for questao in (gabarito.keys()):
        while True:
            resposta = input(f"Sua resposta para a questão {questao}: ").upper()
            if resposta in ['A', 'B', 'C', 'D']:
                respostas_aluno[questao] = resposta
                break
            else:
                print("Resposta inválida. Por favor, digite A, B, C ou D.")
    print(respostas_aluno)
    for questao, resposta_correta in gabarito.items():
        if respostas_aluno[questao] == resposta_correta:
            pontuacao += 1

    print("\n--- Resultado Final ---")
    print(f"Você acertou {pontuacao} de {len(gabarito)} questões.")
    print(f"Sua nota é: {pontuacao}")

corrigir_respostas()
```
