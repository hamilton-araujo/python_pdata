```python
import numpy as np
```

```python
dados = np.loadtxt("citrus.csv", delimiter=",", usecols=np.arange(1,5,1), skiprows=1)
```

```python
peso_laranja = dados[0:5000,1]
diametro_laranja = dados[0:5000,0]
```

```python
peso_toranja = dados[5001:,1]
diametro_toranja = dados[5001:,0]
```

```python
import matplotlib.pyplot as plt

plt.plot(peso_laranja,diametro_laranja)
plt.plot(peso_toranja, diametro_toranja)
```

```python
n = np.size(peso_laranja)
a = (n*np.sum(np.multiply(diametro_laranja,peso_laranja)) - (np.sum(diametro_laranja)*np.sum(peso_laranja)))/(n*np.sum(np.power(peso_laranja,2)) - np.power(np.sum(peso_laranja),2))
b = np.mean(diametro_laranja) - a*np.mean(peso_laranja)

plt.plot(peso_laranja, diametro_laranja)
plt.plot(peso_laranja, a*peso_laranja+b)
```

```python
n = np.size(peso_toranja)
a = (n*np.sum(np.multiply(diametro_toranja,peso_toranja)) - (np.sum(diametro_toranja)*np.sum(peso_toranja)))/(n*np.sum(np.power(peso_toranja,2)) - np.power(np.sum(peso_toranja),2))
b = np.mean(diametro_toranja) - a*np.mean(peso_toranja)

plt.plot(peso_toranja, diametro_toranja)
plt.plot(peso_toranja, a*peso_toranja+b)
```
